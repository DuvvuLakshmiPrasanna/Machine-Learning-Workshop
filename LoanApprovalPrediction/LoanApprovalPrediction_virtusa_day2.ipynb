```python
import pandas as pd
```

```python
# data=pd.read_csv(r"C:\Users\prasa\Downloads\LoanApprovalPrediction.csv")
data =pd.read_csv("LoanApprovalPrediction.csv")
```

```python
#print(data.describe())
print(data.info())
#print(data.columns)
#print(data.size)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python

sample=['Gender', 'Married',  'Education',
       'Self_Employed', 'Property_Area']
```

```python
for i in sample:
    data[i]=le.fit_transform(data[i])
```

```python
data=data.drop(['Loan_ID'],axis=1)
```

```python
data['Dependents']=data['Dependents'].fillna(0)#dropna()
```

```python
data['LoanAmount']=data['LoanAmount'].fillna(1)#dropna()
data['Loan_Amount_Term']=data['Loan_Amount_Term'].fillna(data['Loan_Amount_Term'].mode()[0])#dropna()
data['Credit_History']=data['Credit_History'].fillna(0)#dropna()
```

```python
#####print(data)
x=data.iloc[ : , :-1].values
y=data.iloc[ : ,-1].values
```

```python
##
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=31)
```

```python
from sklearn.linear_model import LogisticRegression
lrmodel=LogisticRegression()
##from sklearn.neighbors import KNeighborsClassifier
##knnmodel=KNeighborsClassifier-(n_neighbors=111
```

```python
lrmodel.fit(xtrain,ytrain)
```

```python
ypred=lrmodel.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score,precision_score,confusion_matrix
print(accuracy_score(ytest,ypred)*100)
print(confusion_matrix(ytest,ypred))
```
