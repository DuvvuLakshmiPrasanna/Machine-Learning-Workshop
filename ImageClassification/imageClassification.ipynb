```python
# Import TensorFlow and Keras modules
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
```

```python
# Load CIFAR-10 dataset
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()
```

```python
# Normalize pixel values from 0-255 to 0-1
train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
# 4. BUILD CNN MODEL
# ==========================================================

# Sequential model means layers are added one after another

model = models.Sequential([

    # ======================================================
    # 4.1 FIRST CONVOLUTION LAYER
    # ======================================================

    # 32 filters
    # Filter size = 3x3
    # ReLU activation
    # Input image size = 32x32 RGB

    layers.Conv2D(
        filters=32,
        kernel_size=(3,3),
        activation='relu',
        input_shape=(32,32,3)
    ),

    # ======================================================
    # 4.2 FIRST MAX POOLING LAYER
    # ======================================================

    # Reduce image size
    # Keeps important features

    layers.MaxPooling2D((2,2)),



    # ======================================================
    # 4.3 SECOND CONVOLUTION LAYER
    # ======================================================

    # Learn more complex features

    layers.Conv2D(
        filters=64,
        kernel_size=(3,3),
        activation='relu'
    ),

    # ======================================================
    # 4.4 SECOND MAX POOLING LAYER
    # ======================================================

    layers.MaxPooling2D((2,2)),


    # ======================================================
    # 4.5 THIRD CONVOLUTION LAYER
    # ======================================================

    # Learn deeper image patterns

    layers.Conv2D(
        filters=64,
        kernel_size=(3,3),
        activation='relu'
    ),


    # ======================================================
    # 4.6 FLATTEN LAYER
    # ======================================================

    # Convert 2D feature maps into 1D vector

    layers.Flatten(),


    # ======================================================
    # 4.7 HIDDEN DENSE LAYER
    # ======================================================

    # Fully Connected Layer

    layers.Dense(
        units=64,
        activation='relu'
    ),


    # ======================================================
    # 4.8 OUTPUT LAYER
    # ======================================================

    # 10 neurons because CIFAR-10 has 10 classes
    # Softmax gives probability of each class

    layers.Dense(
        units=10,
        activation='softmax'
    )
])
```

```python
# 5. COMPILE MODEL
# ==========================================================

# Optimizer = Adam
# Loss Function = Sparse Categorical Crossentropy
# Metric = Accuracy

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# 6. TRAIN MODEL
# ==========================================================

# epochs = number of complete passes through dataset

history = model.fit(
    train_images,
    train_labels,
    epochs=10,
    validation_data=(test_images, test_labels)
)

```

```python
# 7. EVALUATE MODEL
# ==========================================================

# Check performance on test data

test_loss, test_accuracy = model.evaluate(
    test_images,
    test_labels
)

print("Test Accuracy :", test_accuracy)
```

```python
# 8. PREDICT NEW IMAGE
# ==========================================================

# Predict first test image

prediction = model.predict(test_images[:1])

# Get class with highest probability

predicted_class = prediction.argmax()

print("Predicted Class :", predicted_class)
```

```python
# 9. DISPLAY MODEL SUMMARY
# ==========================================================

# Shows all layers and parameters

model.summary()
```

```python
# ==========================================================
# UPLOAD IMAGE IN KAGGLE AND PREDICT
# ==========================================================

from IPython.display import display
from ipywidgets import FileUpload
from tensorflow.keras.preprocessing import image
import numpy as np

# CIFAR-10 class names
class_names = [
    'Airplane',
    'Automobile',
    'Bird',
    'Cat',
    'Deer',
    'Dog',
    'Frog',
    'Horse',
    'Ship',
    'Truck'
]

# Upload widget
uploader = FileUpload(accept='image/*', multiple=False)

display(uploader)

print("Upload an image, then run the next cell.")
```

```python
# Get uploaded file

filename = list(uploader.value.keys())[0]

# Save uploaded image
with open(filename, "wb") as f:
    f.write(uploader.value[filename]["content"])

# Load image
img = image.load_img(filename, target_size=(32,32))

# Convert to array
img_array = image.img_to_array(img)

# Normalize
img_array = img_array / 255.0

# Add batch dimension
img_array = np.expand_dims(img_array, axis=0)

# Predict
prediction = model.predict(img_array)

predicted_class = np.argmax(prediction)

print("Predicted Class:", class_names[predicted_class])
print("Confidence:", round(np.max(prediction) * 100, 2), "%")
```

```python
# ==========================================================
# 10. SAVE MODEL ARCHITECTURE + WEIGHTS
# ==========================================================

# Save model architecture as JSON
model_json = model.to_json()

with open("cnn_cifar10_model.json", "w") as json_file:
    json_file.write(model_json)

# Save weights
model.save_weights("cnn_cifar10_weights.weights.h5")

print("Model architecture and weights saved successfully!")
```
