```python
data =pd.read_csv("Housing.csv")
###Data Understanding
#print(data.describe())
print(data.info())
#print(data.columns)
#print(data.size)
```

```python
###Data Preprocessing: while training data , if you get any errors, implement them here
###To convert strings into numbers(LabelEncoder)
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
###independent variables, Price is Target variable
sample=['mainroad','guestroom', 'basement',
'hotwaterheating', 'airconditioning', 'prefarea', 'furnishingstatus']
```

```python
###Fit is to train the data
for col in sample:
    data[col] = le.fit_transform(data[col])
```

```python
#print(data)
###intergerLocation
###[row(start:end),column(start:end)]
x=data.iloc[ : ,1: ].values
y=data['price'].values
```

```python
###Spliting of trainingdata and testingdata(Model Selection)
###testSize=testingData, Random_State=shuffle
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=198)
```

```python
###Algorithm
###Default n_neighbors=5
from sklearn.neighbors import KNeighborsRegressor
knnmodel=KNeighborsRegressor(n_neighbors=100)

knnmodel.fit(xtrain,ytrain)
```

```python
###predictionResult
ypred=knnmodel.predict(xtest)
```

```python
###To remove negative values(square the values, for the Actual total sum -->squareroot of it ==Actual error)
from sklearn.metrics import mean_squared_error
import math
print(math.sqrt(mean_squared_error(ytest,ypred)))
```
